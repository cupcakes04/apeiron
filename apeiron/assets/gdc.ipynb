````python
import requests
import os
import pandas as pd
from pathlib import Path
from pprint import pprint

# Constants
GDC_API = "https://api.gdc.cancer.gov/files"
DOWNLOAD_API = "https://api.gdc.cancer.gov/data"
ROOT = '/mnt/a/prototype/DATASETS'
OUT_ROOT = '/mnt/d/TEMP_DATA'

task = "mutation"

if task == "mutation":
    LABEL_CSV = ROOT / "task16_gene_mutation/FLEX/Dataset/TCGA-BRCA/tcga-brca_label.csv"
    STATUS_CSV_FILE = ROOT / "breast/download_status.csv"
    OUTPUT_DOWNLOAD_DIR = OUT_ROOT / "task16_gene_mutation"
    label_csv = pd.read_csv(LABEL_CSV)
    submitter_id = label_csv['case_id'].tolist()
    file_id = label_csv['slide_id'].tolist()
    field_value = [
        ("files.data_format", ["SVS"]),
        ("cases.submitter_id", submitter_id), 
        # ("files.file_id", file_id)
    ]
    
if task == "tcga-ot":
    LABEL_CSV = ROOT / "task5_tcga_ot/tcga-ot_val.csv"
    STATUS_CSV_FILE = ROOT / "task5_tcga_ot/download_status.csv"
    OUTPUT_DOWNLOAD_DIR = ROOT / "task5_tcga_ot/DATA"
    label_csv = pd.read_csv(LABEL_CSV)
    submitter_id = label_csv['case_id'].tolist()
    project_id = label_csv['project_id'].tolist()
    field_value = [
        ("files.data_format", ["SVS"]),
        ("cases.submitter_id", submitter_id), 
        ("cases.project.project_id", project_id)
    ]
        
if task == "tcga-brca":
    LABEL_CSV = ROOT / "TCGA_BRCA_Filtered/labels.csv"
    STATUS_CSV_FILE = ROOT / "TCGA_BRCA_Filtered/download_status.csv"
    OUTPUT_DOWNLOAD_DIR = ROOT / "TCGA_BRCA_Filtered/DATA"
    field_value = [
        ("files.data_format", ["SVS"]),
        ("cases.project.project_id", ["TCGA-BRCA"]),
        ("files.experimental_strategy", ["Diagnostic Slide"]),
    ]
    

if task == "tcga-stad":
    LABEL_CSV = ROOT / "task6_msi_gastric/labels.csv"
    STATUS_CSV_FILE = ROOT / "task6_msi_gastric/download_status.csv"
    OUTPUT_DOWNLOAD_DIR = ROOT / "task6_msi_gastric/DATA"
    field_value = [
        ("files.data_format", ["SVS"]),
        ("cases.project.project_id", ["TCGA-STAD"]),
        ("files.experimental_strategy", ["Diagnostic Slide"]),
    ]
    
if task == "tcga-prost":
    LABEL_CSV = ROOT / "Prostate/labels.csv"
    STATUS_CSV_FILE = ROOT / "Prostate/download_status.csv"
    OUTPUT_DOWNLOAD_DIR = ROOT / "Prostate/DATA"
    label_csv = pd.read_csv(LABEL_CSV)
    file_id = label_csv['slide_id'].tolist()
    field_value = [
        ("files.data_format", ["SVS"]),
        ("files.file_id", file_id),
        ("files.experimental_strategy", ["Diagnostic Slide"]),
    ]
        
def search_gdc_data(field_value, size=1000):
    # Create the output directory if it doesn't exist
    OUTPUT_DOWNLOAD_DIR.mkdir(parents=True, exist_ok=True)

    # Prepare the filters for the GDC API
    filters = {
        "op": "and",
        "content": []
    }
    for field, value in field_value:
        filters["content"].append({
            "op": "in",
            "content": {
                "field": field,
                "value": value
            }
        })

    # Query the GDC API to get the file IDs
    params = {
        "filters": filters,
        "format": "JSON",
        "fields": "",
        "size": str(size)
    }

    response = requests.post(GDC_API, json=params)
    if response.status_code != 200:
        print(f"Error querying GDC API: {response.status_code}")
        return

    files = response.json()["data"]["hits"]
    if not files:
        print("No files found matching the criteria.")
        return
    
    return files


def download_gdc_data(matched_data, chunk_kb=8):
    
    # Create or load the download status CSV
    if STATUS_CSV_FILE.is_file():
        status_df = pd.read_csv(STATUS_CSV_FILE)
    else:
        status_df = pd.DataFrame(columns=['name', 'id'])
        
    # Download each file
    for idx, (file_name, file_id) in enumerate(matched_data):
        file_name = file_name.split(".")[0]
        
        if file_name in status_df['name'].values:
            print(f"{idx} -> File {file_name} already downloaded.")
            continue

        download_url = f"{DOWNLOAD_API}/{file_id}"
        response = requests.get(download_url, stream=True)

        if response.status_code == 200:
            file_path = OUTPUT_DOWNLOAD_DIR / f"{file_name}.svs"
            print(f"{idx} -> Downloading {file_name} to {file_path}")
            with open(file_path, 'wb') as f:
                for chunk in response.iter_content(chunk_size=chunk_kb*1024):
                    f.write(chunk)
            print(f"{idx} -> Downloaded {file_id} to {file_path}")

            # Update the status CSV
            status_df = pd.concat([status_df, pd.DataFrame([{'name': file_name, 'id': file_id}])], ignore_index=True)
            status_df.to_csv(STATUS_CSV_FILE, index=False)
        else:
            print(f"{idx} -> Failed to download {file_id}: {response.status_code}")
            
            
def clean_wsi_csv(csv_path, col_name='slide_id', new_col_name='WSI-CODE'):
    """Reads a CSV, cleans the 'WSI-CODE' column, and saves the modified file."""
    # Load CSV
    df = pd.read_csv(csv_path)

    # Function to clean filenames
    def clean_wsi_name(filename):
        return filename.split('.')[0]  # Keeps only 'TCGA-...-DX1.svs'
    
    # Create the new cleaned column
    cleaned_data = df[col_name].apply(clean_wsi_name)

    # Insert the new column at the very beginning (index 0)
    df.insert(0, new_col_name, cleaned_data)

    # Save back to the same CSV file
    df.to_csv(csv_path, index=False)
    print(f"Updated CSV saved: {csv_path}")
    
# clean_wsi_csv(LABEL_CSV, 'slide_id')
````

````python
import time
# from nicmail import send_email_alert

def download_with_retries(data, max_retries=5, delay=5, reset_time=600):  # reset_time = 10 minutes (600 seconds)
    retries = 0
    last_attempt_time = time.time()
    while True:
        try:
            download_gdc_data(data)
            retries = 0  # Reset retries if successful
            last_attempt_time = time.time()  # Reset timer
            return  # Exit if successful
        except Exception as e:
            retries += 1
            print(f"Attempt {retries} failed: {e}")
            
            # Check if 10 minutes have passed since the last attempt
            if time.time() - last_attempt_time > reset_time:
                print("No errors in the last 10 minutes. Resetting retries.")
                retries = 1  # Reset to 1 because we're in a failed attempt
                last_attempt_time = time.time()
            if retries >= max_retries:
                print(f"Failed after {max_retries} attempts within 10 minutes.")
                # send_email_alert(subject='tcga download failed', body=f"Failed after {max_retries} attempts within 10 minutes.")
                return

            time.sleep(delay)  # Wait before retrying

# Run the function
download_with_retries(matched_data)
````
